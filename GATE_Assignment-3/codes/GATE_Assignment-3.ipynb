```python
import matplotlib.pyplot as plt

x = [-5,0,5] 
y = [-1,-1, 1]

plt.step(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [-5,0,5] 
y = [0,0, 1]

plt.step(x, y)
plt.title('u(t)')
plt.xlabel('$t$')
plt.ylabel('$u(t)$')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [-5,0,5] 
y = [1,1, 0]

plt.step(x, y)
plt.title('u(-t)')
plt.xlabel('$t$')
plt.ylabel('$u(-t)$')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [-5,0,5] 
y = [-1,-1, 0]

plt.step(x, y)
plt.title('-u(-t)')
plt.xlabel('$t$')
plt.ylabel('$-u(-t)$')
plt.show()
```
